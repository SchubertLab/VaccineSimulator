```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = [
    "Recall@5", "MSE- Mean", "MSE- Clonotype Mean", 
    "Mean Clonotype Activation", "Mean Clonotype Activation Predicted", 
    "MSE for Bottom 25%", "MSE for Top 25%"
]

s1_values = [0.20, 0.0643, 0.0643, -0.0323, -0.0293, 0.0527, 0.1606]
t1_values = [0.20, 0.1477, 0.1477, 0.0856, 0.0623, 0.1390, 0.3231]

# Compute average values
average_values = np.mean([s1_values, t1_values], axis=0)

# Plot
plt.figure(figsize=(10, 6))
plt.barh(categories, average_values, color='skyblue')
plt.xlabel("Average Value")
plt.title("Average Metrics for S1 and T1")
plt.grid(axis="x", linestyle="--", alpha=0.7)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the generated CSV file
file_path = "/Users/juliakuhn/Desktop/uni/WS2425/Cmscb/Project_phase/VaccineSimulator/notebooks/model_notebooks/model_results.csv"
data = pd.read_csv(file_path)
```

```python
# Selecting relevant columns for S1 and T1
s1_columns = [col for col in data.columns if "S1" in col or col == "Model"]
t1_columns = [col for col in data.columns if "T1" in col or col == "Model"]

# Creating separate DataFrames for S1 and T1
s1_df = data[s1_columns].set_index("Model")
t1_df = data[t1_columns].set_index("Model")

# Plotting heatmap for S1
plt.figure(figsize=(12, 8))
sns.heatmap(s1_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Metrics for S1")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()

# Plotting heatmap for T1
plt.figure(figsize=(12, 8))
sns.heatmap(t1_df, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Heatmap of Metrics for T1")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


```
